```python
import cv2
import numpy
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

```python
def resize(imgname, imgdir, quality, save = False):
    img = cv2.imread(imgdir + imgname, cv2.IMREAD_UNCHANGED)
    print("Input shape: ",img.shape)
    newimg = cv2.resize(img, (0,0), fx = quality, fy = quality)
    print("Output shape:",newimg.shape)
    
    if save:
        imgoutname = imgname.split('.')[0]+'_resized.'+imgname.split('.')[1]
        cv2.imwrite(imgdir+imgoutname, newimg)
    
    return newimg

def border(imgname, imgdir, borderval = (255,255,255), size = None, ratio = None, save = False):
    #use either size or ratio, priority goes to size
    #size specifies the exact size, ratio is the desired aspect ratio
    
    img = cv2.imread(imgdir + imgname, cv2.IMREAD_UNCHANGED)
    h,w,c = img.shape
    
    print("Input shape: ",img.shape)    
    if size is not None:
        newh,neww = size
        hdif = (newh - h)//2
        wdif = (neww - w)//2
        
        newimg = cv2.copyMakeBorder(
                         img, 
                         hdif, 
                         hdif, 
                         wdif, 
                         wdif, 
                         cv2.BORDER_CONSTANT, 
                         value=borderval
                      )
        print("Output shape:",newimg.shape)
        imgoutname = imgname.split('.')[0]+'_resized.'+imgname.split('.')[1]
        if save:
            cv2.imwrite(imgdir+imgoutname, newimg)
        return newimg
    
    elif ratio is not None:
        pass
        
def crop(imgname, imgdir, borderval = (255,255,255), size = None, save = False):
    img = cv2.imread(imgdir + imgname, cv2.IMREAD_UNCHANGED)
    h,w,c = img.shape
    
    print("Input shape: ",img.shape)    
    if size is not None:
        newh,neww = size
        hdif = (h - newh)//2
        wdif = (w - neww)//2
        
        newimg = img[hdif:h-hdif, wdif:w-wdif,:]
        
        print("Output shape:",newimg.shape)
        imgoutname = imgname.split('.')[0]+'_resized.'+imgname.split('.')[1]
        if save:
            cv2.imwrite(imgdir+imgoutname, newimg)
        return newimg
    
    elif ratio is not None:
        pass
          
```
